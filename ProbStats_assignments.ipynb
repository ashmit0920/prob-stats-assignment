```r
# Q1

f = function(x,y) {
  (2*(2*x+3*y))/5
}

# checking joint density function
int_result = integral2(f, xmin=0, xmax=1, ymin=0, ymax=1)

# marginal distribution g(x) at x=1
g=function(y){
 f(1,y)
}

integral(g,0,1)

# marginal distribution h(y) at y=0
h=function(x){
f(x,0)
}

integral(h,0,1)

# expected value of g(x,y) = xy
fd=function(x,y){
f(x,y)*x*y
}

integral2(fd, 0, 1, 0, 1)
```
